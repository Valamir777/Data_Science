```python
'''
In this exercise you have a random variable x that belongs to the interval between 0 and 1
and its PDF is proportional to x. Plot the graph of this PDF. 
Do not forget about its proper normalization.
'''
```

```python
import matplotlib.pyplot as plt
import numpy as np


def pdf(var):
    return var


x = np.linspace(0, 1, 100)
pdf_data = pdf(x)
pdf_data /= np.trapz(pdf_data, x)

plt.plot(x, pdf_data)
plt.ylabel("PDF")
plt.xlabel("X")
plt.title("random variable x between [0, 1] and its PDF proportional")
plt.show()
```

```python
'''
Using cross-entropy compare two probability distributions 
Q1 = 0.16, 0.19, 0.17, 0.15, 0.19, 0.14 Q2 = 0.13, 0.11, 0.14, 0.19, 0.21, 0.22 
designed to model rolling the die. Which one approximate the die better?
'''
import numpy as np

Q1 = np.array([0.16, 0.19, 0.17, 0.15, 0.19, 0.14])             # True probability (one-hot)
Q2 = np.array([0.13, 0.11, 0.14, 0.19, 0.21, 0.22])             # Predicted probability

cross_entropy_loss = -np.sum(Q1 * np.log(Q2))
print(cross_entropy_loss)
```
