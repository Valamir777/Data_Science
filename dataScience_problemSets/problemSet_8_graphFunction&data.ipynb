```python
%matplotlib inline
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns

"""
Create a plot with graphs of two functions: y = e-x2 and y = 3x2e-x2 . 
Add x and y axes labels and the legend. Turn on the grid. 
Customize curve colors and styles.
"""
sns.set_theme()
f, ax = plt.subplots(figsize = (13,7))

x = np.linspace(-10, 10, 400)
y1 = np.exp( -1 * x ** 2)
y2 = 3 * x ** 2 * np.exp( -1 * x ** 2)

ax.plot(x, y1, color = 'red', linestyle = '-')
ax.plot(x, y2, color = 'blue', linestyle = '--')       
plt.title('Plot Title')
# Set x-axis label
plt.xlabel('X - Axis')
# Set y-axis label
plt.ylabel('Y - Axis')
ax.legend(loc='upper right', labels=["y = e-x2"," = 3x2e-x2"])
```
