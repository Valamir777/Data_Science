```python
"""
Create 3D surface plot of a function z=x2âˆ’y2 . Add axes labels and color bar.
"""
%matplotlib inline
import re, seaborn as sb
import matplotlib.pyplot as pyplt
from matplotlib import cm
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap
```

```python
#define function
def FUNC_Z(x, y):
    return (x**2 - y**2)

# Create Grid
sb.set_style('darkgrid')

X_VAL = np.linspace(-100, 100, 10)
Y_VAL = np.linspace(-100, 100, 10)

X1, Y1 = np.meshgrid(X_VAL, Y_VAL)

Z1 = FUNC_Z(X1, Y1)

# Create graph
norm = pyplt.Normalize(Z1.min(), Z1.max())
colors = cm.viridis(norm(Z1))
rcount, ccount, _ = colors.shape

#Graph 1
axes = pyplt.axes(projection='3d')
axes.plot_surface(X1, Y1, Z1)

surf = axes.plot_surface(X1, Y1, Z1, rcount=rcount, ccount=ccount,
                       facecolors=colors, shade=True)
surf.set_facecolor((0,0,0,0))

pyplt.show()

#Graph 2
axes = pyplt.axes(projection='3d')
axes.plot_wireframe(X1, Y1, Z1)

surf = axes.plot_surface(X1, Y1, Z1, rcount=rcount, ccount=ccount,
                       facecolors=colors, shade=True)
surf.set_facecolor((0,0,0,0))

pyplt.show()
```
