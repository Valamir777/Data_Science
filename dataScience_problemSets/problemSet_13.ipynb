```python
"""
File "bank_customers.csv"
describes bank clinets. It contains three columns: customer age (CustAge), 
period of relationship with the bank in months (Months), and credit limit (CredLim). 
Plot a histogram for the column Month and compute its mean, median and the mode. 
Show these values on the histogram. Make a second copy of the histogram, compute quartiles and show them on the plot.
"""
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

data=pd.read_csv('data/bank_customers.csv')

ax = sb.histplot(data=data, x="Months", bins=len(np.unique(list(data['Months']))), color='grey', alpha=1).set(title='Customer Relationships', xlabel="Length of Relationship in Months", ylabel='Frequency of Relationship Length')

plt.axvline(round(data.mean()['Months']), color='red', linestyle='-', linewidth=2, label='Mean')
plt.axvline(round(data.median()['Months']), color='green', linestyle='--', linewidth=2, label='Median')
plt.axvline(int(data.mode()['Months']), color='yellow', linestyle=':', linewidth=2, label='Mode')
plt.legend()
plt.show
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

data=pd.read_csv('data/bank_customers.csv')

ax= sb.histplot(data=data, x="Months", bins=len(np.unique(list(data['Months']))), color='grey', alpha=1).set(title='Customer Relationships', xlabel="Length of Relationship in Months", ylabel='Frequency of Relationship Length')

plt.axvline(data['Months'].quantile(0.25), color='red', linestyle='-', linewidth=2, label='25% Quartile')
plt.axvline(data['Months'].quantile(0.50), color='green', linestyle='--', linewidth=2, label='5% Quartile')
plt.axvline(data['Months'].quantile(0.75), color='yellow', linestyle=':', linewidth=2, label='5% Quartile')
plt.legend()
plt.show
```
