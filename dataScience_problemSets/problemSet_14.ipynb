```python
"""
File "bank_customers.csv"
describes bank clinets. It contains three columns: customer age (CustAge), 
period of relationship with the bank in months (Months), and credit limit (CredLim). 
Plot a histogram for the column CustAge and compute its standard deviation and the interquartile range. 
Show them on the plot.
"""
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

data=pd.read_csv('data/bank_customers.csv')
ax = sb.histplot(data=data, x="CustAge", bins=20, color='grey', alpha=1).set(title='Bank Clients', xlabel="Customer Age", ylabel='Frequency')

std_dev = data['CustAge'].std()
q1 = data['CustAge'].quantile(0.25)
q3 = data['CustAge'].quantile(0.75)
iqr = q3 - q1

plt.axvline(data['CustAge'].mean(), color='blue', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(data['CustAge'].median(), color='red', linestyle='dashed', linewidth=2, label='Median')
plt.axvline(data['CustAge'].mode()[0], color='green', linestyle='dashed', linewidth=2, label='Mode')
plt.axvline(data['CustAge'].mean() + std_dev, color='black', linestyle='dashed', linewidth=2, label='Standard Deviation')
plt.axvline(data['CustAge'].mean() - std_dev, color='black', linestyle='dashed', linewidth=2)
plt.axvline(q1, color='gold', linestyle='dashed', linewidth=2, label='Interquartile Range')
plt.axvline(q3, color='gold', linestyle='dashed', linewidth=2)
plt.legend()
```
