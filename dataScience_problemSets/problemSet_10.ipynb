```python
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image

img = Image.open('data/SV1200.PNG')
imgplot = plt.imshow(img)
plt.axis('off')

width, height = img.size
uniqueColors = set()

w, h = img.size
for x in range(w):
    for y in range(h):
        pixel = img.getpixel((x, y))
        uniqueColors.add(pixel)

totalUniqueColors = len(uniqueColors)

plt.title(f'Image Height {height} Image Width {width} Unique Colors {totalUniqueColors}')
plt.show()
```
